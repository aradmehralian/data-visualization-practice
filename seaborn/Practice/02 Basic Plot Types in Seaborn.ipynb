```python
n = 100
data = pd.DataFrame(
    {
        "x": np.random.randn(n),
        "y": np.random.randn(n),
        "category": np.random.choice(list("ABCD"), size=n),
        "continuous": np.random.randn(n),
    }
)
```

```python
data
```

```python
# data-aware behavior

sns.set_theme(style="whitegrid", font_scale=1.2)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

sns.scatterplot(data=data, ax=ax1, x="x", y="y", hue="category", s=80)
ax1.set_title("Discrete Hue (Categorical Data)", fontweight="bold")

sns.scatterplot(data=data, ax=ax2, x="x", y="y", hue="continuous", s=80)
ax2.set_title("Continuous Hue", fontweight="bold")

fig.tight_layout()
```

Seaborn automatically chooses the best color palette based on the type of data. 

### Figure-level vs. Axes-level Functions

```python
# Axes-level function
fig, ax = plt.subplots(figsize=(8, 5))

sns.scatterplot(data=tips, ax=ax, x="total_bill", y="tip", hue="time")
ax.set_title("Axes-level: Scatter Plot", fontweight="bold");
```

```python
# figure-level function
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="time", col="day")
g.figure.suptitle("Figure-level: Faceted Scatter Plots", fontweight="bold");
```

```python
g = sns.displot(data=iris, x="sepal_length", hue="species", kind="kde", fill=True)
g.ax.set_title("Figure-level: KDE Plot", fontweight="bold");
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.boxplot(data=tips, ax=ax, x="day", y="total_bill")
ax.set_title("Customized Axes-level Plot", fontweight="bold")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Total Bill ($)")

fig.tight_layout()
```

```python
g = sns.catplot(data=tips, x="day", y="total_bill", kind="box", height=4, aspect=1.5)
g.set_axis_labels("Day of the Week", "Total Bill ($)")
g.ax.set_title("Customized Figure-level Plot", fontweight="bold")
g.set_xticklabels(rotation=45)

g.figure.tight_layout()
```
