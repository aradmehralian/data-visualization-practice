```python
x = np.linspace(0, 4 * np.pi, 300)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, color="salmon")
ax.set(title="Sine Wave", xlabel="X axis", ylabel="Y axis");
```

```python
sns.set_theme(style="darkgrid")

fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=x, y=y, ax=ax, color="salmon")
ax.set(title="$\sin{x}$", xlabel="X axis", ylabel="Y axis");
```

```python
x = np.random.randint(0, 6, size=8)
y = 2 * x + 5

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# darkgrid style
sns.scatterplot(x=x, y=y, ax=ax1, color="magenta")
ax1.set_xlim(-0.3, 8)
ax1.set_ylim(y.min() - 1, y.max() + 1)
ax1.set_title("Dark Grid Style")
ax1.set_facecolor("#EAEAF2")
ax1.grid(True, color="white", linewidth=1.2)

# whitegrid style
sns.scatterplot(x=[1, 2, 3, 4, 5], y=[2, 4, 6, 8, 10], ax=ax2, color="navy")
ax2.set_title("White Grid Style")
ax2.set_facecolor("white")
ax2.grid(True, color="#DCDCDC", linewidth=1)

fig.tight_layout()
```

```python
x = np.random.normal(loc=0, scale=1, size=100)
y = x * 2 + np.random.normal(0, 2, 100)

fig, ax = plt.subplots(figsize=(8, 6))
sns.regplot(x=x, y=y, ax=ax, color="darkred")
ax.set_title("Regression Plot with Confidence Interval", fontweight="bold");
```

```python
tips = sns.load_dataset("tips")
palettes = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

fig, axes = plt.subplots(2, 3, figsize=(12, 10), sharey=True)
axes = axes.flatten()

for i, palette in enumerate(palettes):
    sns.set_palette(palette)
    sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", ax=axes[i])
    axes[i].set_title(f"{palette.title()} Palette")

fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
font_params = {"fontweight": "bold", "color": "navy", "fontsize": 12}

sns.violinplot(
    data=tips,
    y="total_bill",
    x="day",
    ax=ax,
    inner="points",
    color="salmon"
)

ax.set_title("Violin Plots with Inner Points", **font_params);
```
