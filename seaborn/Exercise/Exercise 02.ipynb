```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme(
    context="notebook", style="white", rc={"figure.figsize": (14, 6)}, palette="tab10"
)
```

```python
df = pd.read_csv("data/spotify.csv", index_col="Date", parse_dates=True)
```

```python
df.head(10)
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.lineplot(data=df, ax=ax)
ax.set_title("Daily Global Streams of Popular Songs in 2017-2018", fontweight="bold")

fig.tight_layout()
```

```python
list(df.columns)
```

```python
ax = sns.lineplot(data=df, x="Date", y="Shape of You", label="Shape of You")
sns.lineplot(data=df, ax=ax, x="Date", y="Despacito", label="Despacito")
ax.set(ylabel=None)
ax.legend(title="Name of Song")

plt.tight_layout()
```
