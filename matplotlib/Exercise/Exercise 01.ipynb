```python
# 2
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(df["month_number"], df["total_profit"], marker="o")
ax.set(title="Company Profit per Month", ylabel="Total Profit", xlabel="Month Number")
ax.set_ylim(150_000, 450_000)

ax.yaxis.set_major_formatter(mtick.StrMethodFormatter("${x:,.0f}"))

fig.tight_layout()
```

```python
# 3
fig, ax = plt.subplots(figsize=(8, 6))

style_params = {
    "color": "red",
    "linestyle": "dashed",
    "linewidth": 3,
    "marker": "o",
    "markerfacecolor": "black",
}

ax.plot(df["month_number"], df["total_units"], label="total units", **style_params)
ax.legend(loc="upper left")
ax.set(title="Company Sales (2024)", xlabel="Month Number", ylabel="Sold Units Number")

ax.yaxis.set_major_formatter(mtick.StrMethodFormatter("{x:,.0f}"))

fig.tight_layout()
```
