## Introduction to `matplotlib`

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
mpl.__version__
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y, color="red")
plt.title("Since Wave")
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 30)
y = np.cos(x)

# Matplotlib's object hierarchy
fig, ax = plt.subplots()

# Artists
ax.plot(x, y, color="orange", marker="o")
ax.set_xlabel("$x$")
ax.set_ylabel("$cos(x)$")
ax.set_title("Cosine Wave")

plt.show()
```

### Line Plot

```python
x = np.linspace(-5, 5, 100)
y = x**2

# create figure and axe objects
fig, ax = plt.subplots(figsize=(8, 6))

# create plot and set labels
ax.plot(x, y, color="purple", linewidth=1.5, linestyle="--")
ax.set(xlabel="X", ylabel="$y = x^2$", title="Quadratic Function")

plt.show()
```

### Multiple Lines

```python
x = np.linspace(0, 2 * np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

# create figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# artists
ax.plot(x, y1, label="$sin(x)$")
ax.plot(x, y2, label="$cos(x)$")

# add label and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Sine and Cosine Functions", fontsize=16)
ax.set_xlim(left=0)

# add legend
ax.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

### Scatter Plot

```python
# random data
np.random.seed(7)
x = np.random.rand(50)
y = np.random.rand(50)


# create plot
fig, ax = plt.subplots(figsize=(9, 7))
ax.scatter(x, y, color="magenta", alpha=0.75)

# add labels, titles, limits
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.set_title("Random Scatter Plot", fontsize=14)
ax.set_xlim([0, 1.1])
ax.set_ylim([0, 1])


plt.tight_layout()
plt.grid()
plt.show()
```
