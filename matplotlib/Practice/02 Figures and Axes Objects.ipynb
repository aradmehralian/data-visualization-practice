### Subplots

```python
x = np.linspace(0, 10, 200)

fig, axs = plt.subplots(2, 2, figsize=(8, 6), sharex=True)
axs[0, 0].plot(x, np.sin(x))
axs[0, 1].plot(x, np.cos(x))
axs[1, 0].plot(x, np.exp(x))
axs[1, 1].plot(x, x**2)

for ax in axs.flat:
    ax.set(xlabel="$x$", ylabel="$f(x)$")
    ax.grid()

fig.tight_layout(pad=0.8)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
fig.tight_layout()
fig.suptitle("Trigonometric Functions")
gs = gridspec.GridSpec(2, 2)  # control the grid of subplots


ax1 = fig.add_subplot(gs[0, 0])  # top left
ax2 = fig.add_subplot(gs[0, 1])  # top right
ax3 = fig.add_subplot(gs[1, :])  # bottom, all columns

ax1.plot(x, np.sin(x), color="blue")
ax2.plot(x, np.cos(x), color="red")
ax3.plot(x, np.tan(x), color="purple")

ax1.set(title="$sin(x)$")
ax2.set(title="$cos(x)$")
ax3.set(title="$tan(x)$")

ax1.grid()
ax2.grid()
ax3.grid()

plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 500)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

ax1.plot(x, np.sin(x))
ax1.set_title("$sin(x)$")
ax2.plot(x, np.cos(x))
ax2.set_title("$cos(x)$")

ax1.grid(True, linestyle="--", alpha=0.5)
ax2.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

### Customizing Figure and Axes

```python
x = np.linspace(1, 6)
y = x

fig, ax = plt.subplots(figsize=(6, 4), dpi=150, facecolor="lightgray")
ax.plot(x, y)
ax.set(title="$y = x$", xlabel="$x$", ylabel="$y$")
ax.set_xlim(1, 5)
ax.set_ylim(1, 4)

ax.grid(True, linestyle="--", linewidth=0.4, alpha=0.6)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [3, 4, 5, 6])
ax.set(
    xticks=[1, 2, 3, 4],
    yticks=[3, 4, 5, 6, 7],
    yticklabels=["Low", "Medium", "Hight", "Very High", "Ultra High"],
)
ax.set_xticklabels(["Zero", "One", "Two", "Three"], rotation=45)

plt.show()
```

### Exercise

```python
# data
x = np.linspace(0, 10, 500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x / 10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y1, label="$sin(x)$", color="navy")
ax.plot(x, y2, label="$cos(x)$", color="firebrick")
ax.plot(x, y3, label="$e^{-x/10}$", color="springgreen")

ax.set(title="Comparison of Functions", xlabel="X-axis", ylabel="Y-axis")
ax.legend(loc="lower left")
ax.grid(True, linestyle="--", linewidth=0.5, alpha=0.7)

plt.show()
```

```python
# data
np.random.seed(21)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 250 * np.random.rand(100)

font_params = {"fontweight": "bold", "font": "monospace", "size": 14, "c": "crimson"}

fig, ax = plt.subplots(figsize=(8, 6))

scatter_plot = ax.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap="viridis")
ax.set_xlabel("X-axis", size=12)
ax.set_ylabel("Y-axis", size=12)
ax.set_title("Scatter Plot with Color Mapping", **font_params)
ax.set_facecolor("seashell")

# add color bar to figure
cbar = fig.colorbar(scatter_plot)
cbar.set_label("Color Value")

fig.tight_layout()
fig.set_facecolor("lavender")
plt.show()
```

```python
# data
categories = ["A", "B", "C", "D", "E"]
values = np.random.randint(15, 40, size=5)
```

```python
# create the plot
fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(categories, values, color="dodgerblue", edgecolor="navy")
ax.set_title("Bar Plot", fontweight="light", fontstyle="italic", size=14)
ax.set_xlabel("Categories", size=12)
ax.set_ylabel("Values", size=12)
ax.set_ylim(0, np.max(values) + 5)
ax.set_facecolor("khaki")

for i, height in enumerate(values):
    ax.text(i, height + 0.5, s=str(height), ha="center", c="darkmagenta")

fig.tight_layout()
fig.set_facecolor("lightgrey")
plt.show()
```
