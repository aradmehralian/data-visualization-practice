## Working with Images

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pathlib
```

```python
img_path = pathlib.Path("./data/Nature.jpg")
img = mpimg.imread(img_path) # reads and saves the image's array
```

```python
img.shape
```

```python
img.dtype
```

```python
type(img)
```

```python
plt.imshow(img); # implicit interface
```

```python
# explicit interface
fig, ax = plt.subplots(figsize=(10, 8))
ax.imshow(img);
ax.set_axis_off()
```

```python
new_img = img[:, :, 0] # getting blue channel
```

```python
fig, ax = plt.subplots(figsize=(9, 7))
ax.imshow(new_img, cmap="hot");
ax.set_axis_off()
```

```python
# getting figure and axes objects
fig, ax = plt.subplots(figsize=(8.5, 6))

# plotting the image
picture = ax.imshow(new_img, cmap="coolwarm")
ax.set_axis_off() # turn axis off

# adding colorbar
fig.colorbar(picture, label="Color Value")

fig.set_facecolor("lightgrey")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 12))

title_props = {
    "fontweight": "bold",
    "fontfamily": "Lato",
    "size": 14,
}

ax1.imshow(img, interpolation="hamming")
ax1.set_title("Hamming Interpolation", **title_props)
ax1.set_axis_off()

ax2.imshow(img, interpolation="nearest")
ax2.set_title("nearest interpolation".title(), **title_props)
ax2.set_axis_off()


fig.tight_layout();
```
