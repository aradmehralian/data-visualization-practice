## Basic Plot Types with `matplotlib`

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Line Plots

```python
# data
x = np.linspace(0, 50, 500)
y = np.sqrt(x)

# figure and axes objects
fig, ax = plt.subplots(figsize=(6, 4))

# plot the data
ax.plot(x, y, color="orange")

# title and labels
ax.set(title="Square Root Function", xlabel="$X$", ylabel="$\sqrt{X}$")

plt.show()
```

```python
# data
x = np.linspace(0, 10, 300)

# plot data
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(x, np.sin(x), label="$sin(x)$")
ax.plot(x, np.cos(x), label="$cos(x)$")
ax.plot(x, np.sin(x) + 2, label="$sin(x) + 2$", linestyle="-.", linewidth=2, color="darkmagenta")

# customize axis
ax.set_xlim(0, 2*np.pi)
ax.set_title("Trigonometric Functions", fontweight="bold")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# add legend, grid
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5)

fig.tight_layout()
plt.show()
```

### Scatter Plot

```python
# data
np.random.seed(28)
x = np.random.rand(100)
y = x + np.random.normal(5, 1.5, 100)

# figure and axes objects
fig, ax = plt.subplots()
ax.scatter(x, y)

# title and labels
ax.set_title("Basic Scatter Plot", fontstyle="italic", fontsize=16)
ax.set(xlabel="X-axis", ylabel="Y-axis")


plt.grid(True, alpha=0.5)
plt.show()
```

```python
# data
np.random.seed(39)
x = np.random.rand(50)
y = x + np.random.normal(3, 0.25, size=50) * 0.8
colors = np.random.rand(50)
sizes = np.random.rand(50) * 300

# figure and axes objects
fig, ax = plt.subplots(figsize=(8, 6))

# create scatter plot
scatter = ax.scatter(x, y, s=sizes, c=colors, cmap="cool")

# add color bar
cbar = fig.colorbar(scatter)
cbar.set_label("Color Value", size=12)

# customize the plot
ax.set_title("Customized Scatter Plot", fontsize=14, fontweight="bold", c="midnightblue")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.grid(True, linestyle="--", linewidth=0.7, alpha=0.5)

# add trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax.plot(x, p(x), "r", alpha=0.7, label="Trend Line")

ax.legend()
fig.tight_layout(pad=3)
plt.show()
```

### Bar Charts

```python
# create temperature data
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
values = np.random.randint(20, 30, size=7)

# figure and axes objects
fig, ax = plt.subplots(figsize=(8, 6))

# create bar chart
ax.bar(days, values, color="skyblue", edgecolor="black")
ax.set_title("Basic Bar Chart", fontsize=14)
ax.set_ylabel("Temperatures ($C^{o}$)")
ax.grid(True, axis="y", linestyle="--", alpha=0.5)

# rotate x-tick labels
for label in ax.get_xticklabels():
    label.set_rotation(45)

# Add value labels to the each bar
for i, v in enumerate(values):
    ax.text(i, v, str(v), ha="center", va="bottom")

fig.tight_layout()
plt.show()
```

### Histograms

```python
data = np.random.normal(loc=100, scale=20, size=1000)

# fig and axes objects
fig, ax = plt.subplots(figsize=(8, 4))

# create histogram
ax.hist(data, bins=30, edgecolor="black", color="salmon")
ax.set_title("Basic Histogram", c="orangered", fontweight="bold")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")

# add vertical line for mean
ax.axvline(x=data.mean(), color="navy", linestyle="dashed", label="Mean")


ax.legend()
plt.show()
```
